## Isolation Forest

Dalam metode ini kuatifikasi jumlah outlier ditentukan dengan menggunakan interquartile range yang mana dapat divisualisasikan dengan bocplot outlier tersebut

```python
#Memplot box plot pada dataset
plt.figure(figsize=(20,5))
sns.boxplot(data=md,color='blue')
```

Jumlah outlier tersebut dihitung menggunakan persamaan interquarile dengan skala 1,5

```python
#Menginisialisasi perhitungan outlier dengan metode quartile
Q1 = md.quantile(0.25)
Q3 = md.quantile(0.75)
IQR = Q3 - Q1
```

```python
# Menentukan banyaknya outlier pada setiap fitur dengan metode outlier
banyak_setiap_outlier = (((md < (Q1 - 1.5 * IQR)) | (md > (Q3 + 1.5 * IQR))).sum())
banyak_setiap_outlier
```

```python
#Menentukan banyaknya outlier pada sistem
outlier=max(((md < (Q1 - 1.5 * IQR)) | (md > (Q3 + 1.5 * IQR))).sum())
print("Jumlah outlier :" , outlier)
```

```python
#Menghitung jumlah seluruh data
jumlah_data=max(md.count())
print("Jumlah seluruh data:", jumlah_data)
```

```python
#Menghitung fraksi kontaminasi 
contamination_value=(outlier)/(jumlah_data)
print("Fraksi contaminasi: ", contamination_value)
```

```python
#Menggunakan metode isolation forest
model=IsolationForest(n_estimators=100, max_samples='auto', contamination=float(contamination_value) ,max_features=1.0,verbose=1)
model.fit(md[['Bearing 1', 'Bearing 2', 'Bearing 3', 'Bearing 4']])
```

Nilai score pada label ini menunjukan seberapa dekat suatu data dapat dikategorikan anomali ataupun normal.
semakin mendekati scores -0.5. maka data tersebut semakin jauh dari kumpulan normal. Untuk data yang bernilai positif diatas 0 maka data tersebut. Semakin mendekati 0.5 maka data tersebut semakin kuat untuk dikatakan normal.

```python
# Memunculkan score dan anomaly
md['scores']=model.decision_function(md[['Bearing 1', 'Bearing 2', 'Bearing 3', 'Bearing 4']])
md['anomaly']=model.predict(md[['Bearing 1', 'Bearing 2', 'Bearing 3', 'Bearing 4']])
md.head(20)
```
